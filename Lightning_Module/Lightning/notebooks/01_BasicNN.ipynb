```python
%load_ext autoreload
%autoreload 2

import Lightning.models.basic_nn as basic_nn
import torch
import matplotlib.pyplot as plt
import seaborn as sns

import Lightning.script.train as train
```

```python
input_doses = torch.linspace(start=0, end=1, steps=11)
```

```python
model = basic_nn.BasicNN()
output_values = model(input_doses)
```

```python
sns.set_style("whitegrid")
sns.lineplot(x=input_doses, y=output_values, color='green', linewidth=2.5)
plt.ylabel("Effectiveness")
plt.xlabel("Dose")
```

```python
model_train = basic_nn.BasicNN_train()
output_values = model_train(input_doses)

sns.set_style("whitegrid")
# We call detach() on the output values to create a new tensor that only has the values
sns.lineplot(x=input_doses, y=output_values.detach(), color='green', linewidth=2.5)
plt.ylabel("Effectiveness")
plt.xlabel("Dose")
```

```python
inputs = torch.tensor([0, 0.5, 1.])
labels = torch.tensor([0., 1., 0.])
```

```python
train.train(model_train, inputs, labels)
```

```python
output_values = model_train(input_doses)

sns.set_style("whitegrid")
# We call detach() on the output values to create a new tensor that only has the values
sns.lineplot(x=input_doses, y=output_values.detach(), color='green', linewidth=2.5)
plt.ylabel("Effectiveness")
plt.xlabel("Dose")
```
