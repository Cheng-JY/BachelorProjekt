```python
import numpy as np

from sklearn.metrics import pairwise_distances
```

```python
X = np.random.rand(100,3)
```

```python
cluster_idx = [1,10,99]
clusters = X[cluster_idx]
dist_matrix = pairwise_distances(X, clusters, metric="euclidean")
print(dist_matrix[:3, :])
dist = np.min(dist_matrix, axis=1).reshape(-1,1)
print(dist)
```

```python
random_distance = np.random.rand(100,1)
print(random_distance[:3, :])
```

```python
distance = np.minimum(random_distance, dist_matrix)
print(distance)
```
